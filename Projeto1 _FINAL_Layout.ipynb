# Renda

Nesse campo entram a relação entre o acesso a TV e Internet e a renda que ganham em 2015. Nisso eu observo se houve aumento da renda e se por isso hove aumento desse acesso, ou se é o oposto.

# Acre

```python
ren2015 = At2015.Renda_per_capta_domiciliar
ren2015.describe()
```

```python
plot = ren2015.plot(kind='box', figsize = (8,8),title='Renda Percapta Domiciliar Acre 2015')#, color='red')
plt.ylabel(('Renda per capta domiciliar Acre(em reais)'))
```

Acre: Utilizando o conceito de amplitude interquartil, definimos que ualquer vaolor que não esteja no intervalo [Q1 - 1,5Q: Q3 + 1,5Q] sendo Q = Q3 - Q1, Q3 (Terceiro Quartil) e Q1 (Primeiro Quartil), ou seja [-800:2080] seria extremo.

```python
faixa = (0,937/4 ,937/2, 937, 937*2,2080)
ren2015rel = pd.cut(ren2015, faixa, False)
print('Frequências relativas:')
ut1 = (ren2015rel.value_counts(sort=False, normalize=True)*100).round(decimals=1)
print(ut1,'\n')
```

```python
renda_Internet2015=pd.crosstab( ren2015rel, At2015.Acesso_Internet, margins=True, rownames=['Renda per capta domiciliar Acre (em salários mínimos)'], colnames=['Acesso Internet no Acre'], normalize=True)*100
renda_Internet2015=renda_Internet2015.round(decimals=1)
renda_Internet2015
```

```python
renda_TV2015=pd.crosstab( ren2015rel, At2015.Tem_TV, margins=True, rownames=['Renda per capta domiciliar Acre (em salários mínimos)'], colnames=['Tem Tv Acre'], normalize=True)*100
renda_TV2015=renda_TV2015.round(decimals=1)
renda_TV2015
```
